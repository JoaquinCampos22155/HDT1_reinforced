# 1. Elija un sistema real de su ГЎrea de interГ©s
## Sistema elegido

**Sistema de optimizaciГіn de campaГұas de marketing digital**, donde una IA analiza el rendimiento de las campaГұas y decide cГіmo distribuir el presupuesto de manera automГЎtica.

### (a) Estado $S_t$

El estado representa toda la informaciГіn disponible para tomar una decisiГіn en un instante determinado, por ejemplo:

- Presupuesto disponible.
- CTR (Click Through Rate).
- CPC (Cost Per Click).
- NГәmero de conversiones.
- ROAS (Return on Ad Spend).
- Rendimiento histГіrico de cada campaГұa.

Mientras mГЎs completo sea el estado, mejores decisiones podrГЎ tomar el agente.

### (b) Espacio de acciones $\mathcal{A}$

Las acciones posibles incluyen:

- Aumentar el presupuesto de una campaГұa.
- Reducir el presupuesto.
- Pausar una campaГұa.
- Reactivar una campaГұa.
- Cambiar la estrategia de puja.
- Redistribuir presupuesto entre campaГұas.

### (c) FunciГіn de recompensa $R$

La recompensa debe reflejar los objetivos del negocio.

Se puede:

- Recompensar cuando aumentan las conversiones.
- Recompensar cuando mejora el ROAS.
- Penalizar cuando se consume mucho presupuesto sin obtener buenos resultados.

El sistema aprende a optimizar la inversiГіn y no Гәnicamente a conseguir mГЎs clics.

### (d) Observabilidad del entorno

El entorno es **parcialmente observable**.

Aunque el sistema dispone de gran cantidad de datos (CTR, CPC, conversiones, etc.), nunca conoce toda la informaciГіn relevante. Por ejemplo:

- No sabe exactamente quГ© piensa cada usuario.
- No conoce si una persona comprarГЎ mГЎs adelante.
- No puede observar la influencia de anuncios vistos en otras plataformas.

# 2. Valor apropiado de $\gamma$

Para este sistema usarГӯamos:


$\gamma$ = 0.95


porque en marketing digital muchas decisiones producen resultados despuГ©s de cierto tiempo. Una campaГұa nueva puede requerir varios dГӯas para encontrar la audiencia adecuada antes de generar conversiones rentables.

## Ejemplo


- **OpciГіn A:** recompensa inmediata de **+30**.
- **OpciГіn B:** pГ©rdida inicial de **вҲ’10**, seguida de una recompensa futura de **+100**.

### Con $\gamma = 0.95$

\[
-10 + 0.95(100) = 85
\]

El agente considera que vale la pena asumir la pГ©rdida inicial porque la recompensa futura es mucho mayor.

### Con $\gamma = 0.01$

\[
-10 + 0.01(100) = -9
\]

En este caso el agente prГЎcticamente ignora la recompensa futura y evita invertir en campaГұas nuevas, aunque sean mucho mГЎs rentables a largo plazo.

# 3. PolГӯtica determinista vs. polГӯtica estocГЎstica

## PolГӯtica determinista

Una polГӯtica determinista siempre toma la misma decisiГіn para un mismo estado.

Ejemplo:

- Si el **ROAS > 4**, aumentar el presupuesto un **20%**.
- Si el **ROAS < 2**, reducir el presupuesto.

Ante las mismas condiciones, la decisiГіn siempre serГЎ la misma.

## PolГӯtica estocГЎstica

Una polГӯtica estocГЎstica asigna probabilidades a las acciones posibles.

Por ejemplo:

- **60%** de probabilidad de aumentar el presupuesto.
- **25%** de mantenerlo igual.
- **15%** de probar una nueva audiencia.

No siempre ejecuta la misma acciГіn, incluso cuando el estado es idГ©ntico.

## ВҝCuГЎndo preferir una polГӯtica estocГЎstica?

PreferirГӯa una polГӯtica estocГЎstica cuando el objetivo es **explorar nuevas estrategias**.

Si el sistema siempre toma la misma decisiГіn, puede quedarse explotando Гәnicamente las campaГұas que ya funcionan y perder oportunidades de descubrir configuraciones o audiencias que generen un mejor rendimiento en el largo plazo.

## Programa

````python
"""
Entorno: GridWorld 4x4 simple
- El agente parte de (0,0) y busca llegar a (3,3)
- Recompensa: -1 por cada paso, +10 al llegar a la meta, -5 si cae en celda trampa
- El agente sigue una polГӯtica aleatoria (no aprende todavГӯa вҖ” solo observamos el ciclo)
"""

import numpy as np
import random

# Constantes
GRID_SIZE = 4
GOAL_STATE = (3, 3)          # Estado terminal con recompensa positiva
TRAP_STATES = {(1, 1), (2, 3)}  # Estados trampa
MAX_STEPS = 50               # Horizonte mГЎximo del episodio
GAMMA = 0.99                 # Factor de descuento

# Acciones
# Espacio de acciones discreto: 4 direcciones posibles
ACTIONS = {
    0: (-1, 0),   # Arriba
    1: (1, 0),    # Abajo
    2: (0, -1),   # Izquierda
    3: (0, 1),    # Derecha
}
ACTION_NAMES = {0: "вҶ‘", 1: "вҶ“", 2: "вҶҗ", 3: "вҶ’"}

# Entorno
class GridWorldEnv:
    """
    Representa el entorno (Environment) en el vocabulario de RL.
    
    Responsabilidades:
    - Mantener el estado actual del mundo
    - Transicionar al nuevo estado dado una acciГіn
    - Emitir la recompensa R_{t+1} despuГ©s de cada transiciГіn
    """
    
    def reset(self):
        """Inicializa el entorno y devuelve el estado inicial S_0."""
        self.state = (0, 0)
        self.done = False
        return self.state
    
    def step(self, action):
        """
        Ejecuta un paso del ciclo agente-entorno.
        
        Recibe: A_t (action)
        Devuelve: S_{t+1}, R_{t+1}, done
        
        Esta funciГіn implementa la funciГіn de transiciГіn P(S_{t+1} | S_t, A_t)
        y la funciГіn de recompensa R(S_t, A_t, S_{t+1}).
        """
        if self.done:
            raise RuntimeError("El episodio ya terminГі. Llama reset().")
        
        dr, dc = ACTIONS[action]
        r, c = self.state
        
        # La transiciГіn respeta los bordes del grid (entorno determinista)
        new_r = max(0, min(GRID_SIZE - 1, r + dr))
        new_c = max(0, min(GRID_SIZE - 1, c + dc))
        next_state = (new_r, new_c)
        
        # FunciГіn de recompensa: seГұal de diseГұo del ingeniero
        if next_state == GOAL_STATE:
            reward = +10.0    # Meta alcanzada
            self.done = True
        elif next_state in TRAP_STATES:
            reward = -5.0     # Trampa activada
            self.done = True
        else:
            reward = -1.0     # Costo por paso (incentiva llegar rГЎpido)
        
        self.state = next_state
        return next_state, reward, self.done


# Agente con politica aleatoria
class RandomAgent:
    """
    Agente con polГӯtica estocГЎstica uniforme.
    
    PolГӯtica ПҖ(a|s) = 1/|A| para todo a вҖ” distribuciГіn uniforme.
    No aprende вҖ” sirve para ilustrar el ciclo de interacciГіn.
    """
    
    def __init__(self, action_space_size):
        self.n_actions = action_space_size
    
    def select_action(self, state):
        """
        Muestrea una acciГіn de la polГӯtica estocГЎstica uniforme.
        
        En un agente que aprende (Q-learning, PPO, etc.), esta funciГіn
        consultarГӯa una tabla Q o una red neuronal.
        """
        return random.randint(0, self.n_actions - 1)


# Ciclo de agnte entorno
def run_episode(env, agent, gamma=GAMMA, max_steps=MAX_STEPS, verbose=True):
    """
    Ejecuta un episodio completo y calcula el retorno G_0.
    
    Implementa directamente el ciclo de la Diapositiva 3:
        S_0 вҶ’ A_0 вҶ’ R_1, S_1 вҶ’ A_1 вҶ’ R_2, S_2 вҶ’ ...
    """
    state = env.reset()                    # S_0
    trajectory = []                        # Almacena (S_t, A_t, R_{t+1})
    total_reward = 0.0
    
    if verbose:
        print(f"\n{'в•җ'*50}")
        print(f"INICIO DEL EPISODIO  Estado inicial: {state}")
        print(f"{'в”Җ'*50}")
    
    for t in range(max_steps):
        action = agent.select_action(state)                    # A_t ~ ПҖ(В·|S_t)
        next_state, reward, done = env.step(action)            # TransiciГіn
        
        trajectory.append((state, action, reward, next_state))
        total_reward += reward
        
        if verbose:
            status = "вң“ META" if next_state == GOAL_STATE \
                     else "вң— TRAMPA" if next_state in TRAP_STATES \
                     else ""
            print(f"  t={t:2d}  S={state}  A={ACTION_NAMES[action]}  "
                  f"R={reward:+.1f}  S'={next_state}  {status}")
        
        state = next_state
        
        if done:
            break
    
    # Calcula el retorno descontado G_0 = ОЈ Оі^k * R_{k+1}
    # Procesa la trayectoria en reversa para acumulaciГіn eficiente
    G = 0.0
    returns = []
    for _, _, r, _ in reversed(trajectory):
        G = r + gamma * G                  # G_t = R_{t+1} + Оі * G_{t+1}
        returns.insert(0, G)
    
    if verbose:
        print(f"{'в”Җ'*50}")
        print(f"  Pasos totales        : {len(trajectory)}")
        print(f"  Recompensa acumulada : {total_reward:+.1f}")
        print(f"  Retorno G_0 (Оі={gamma}): {returns[0]:.4f}")
        print(f"{'в•җ'*50}\n")
    
    return trajectory, returns


# Experimentar con multiples episodios
def run_experiment(n_episodes=5):
    """
    Corre mГәltiples episodios para ilustrar la variabilidad de una polГӯtica
    estocГЎstica вҖ” con la misma polГӯtica, los resultados varГӯan por azar.
    """
    env = GridWorldEnv()
    agent = RandomAgent(action_space_size=len(ACTIONS))
    
    print("SIMULACIГ“N: Ciclo Agente-Entorno en GridWorld 4Г—4")
    print("PolГӯtica: aleatoria uniforme (sin aprendizaje)")
    print(f"Factor de descuento Оі = {GAMMA}")
    
    all_returns = []
    for ep in range(n_episodes):
        print(f"\nв–¶ Episodio {ep + 1}/{n_episodes}")
        _, returns = run_episode(env, agent, verbose=True)
        all_returns.append(returns[0])
    
    print("\n" + "в•җ"*50)
    print("RESUMEN DEL EXPERIMENTO")
    print(f"  Retorno promedio G_0 : {np.mean(all_returns):.4f}")
    print(f"  Retorno mГЎximo       : {max(all_returns):.4f}")
    print(f"  Retorno mГӯnimo       : {min(all_returns):.4f}")
    print("в•җ"*50)
    print("\nObservaciГіn: la variabilidad entre episodios ilustra por quГ©")
    print("el objetivo de RL es el RETORNO ESPERADO, no un retorno fijo.")


if __name__ == "__main__":
    random.seed(42)
    np.random.seed(42)
    run_experiment(n_episodes=5)
````
